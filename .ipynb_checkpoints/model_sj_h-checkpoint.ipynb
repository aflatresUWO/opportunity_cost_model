```julia
using LinearAlgebra
using Plots
using LaTeXStrings
using Profile
using Random
using CSV
using DataFrames
```

```julia
#survivals:
s_u = 0.4
s_x = 0.5
s_y = 0.6
s_jx = 0.4
s_jy = 0.4


#fecundity
bx_0 = 2
by_0 = 4
b_x(s_jx) = bx_0 + s_jx*2
b_y(s_jy) = by_0 + s_jy*2
phi = 0.1
M = 1

#altruism
h_x = 0
h_y = 0

h_x_bar = h_x
h_y_bar = h_y

T_x(h_x_bar) = 1-exp(-b_x(s_jx)*h_x_bar)
T_y(h_y_bar) = exp(-b_y(s_jy)*h_y_bar)

theta_x(eq) = (1-phi)*eq[2]/(eq[2]+eq[3]+s_u*p_u_tilde(eq[2],eq[3])*eq[1]+(1-h_x)*s_u*p_u_tilde(eq[2],eq[3])*eq[4]+h_x*(T_x(h_x_bar)*s_jy+(1-T_x(h_x_bar))*s_jx)*p_j_tilde(eq[2],eq[3])*eq[4]+(1-h_y)*s_u*p_u_tilde(eq[2],eq[3])*eq[5]+h_y*(T_y(h_y_bar)*s_jx+(1-T_y(h_y_bar))*s_jy)*p_j_tilde(eq[2],eq[3])*eq[5])
theta_y(eq) = (1-phi)*eq[2]/(eq[2]+eq[3]+s_u*p_u_tilde(eq[2],eq[3])*eq[1]+(1-h_x)*s_u*p_u_tilde(eq[2],eq[3])*eq[4]+h_x*(T_x(h_x_bar)*s_jy+(1-T_x(h_x_bar))*s_jx)*p_j_tilde(eq[2],eq[3])*eq[4]+(1-h_y)*s_u*p_u_tilde(eq[2],eq[3])*eq[5]+h_y*(T_y(h_y_bar)*s_jx+(1-T_y(h_y_bar))*s_jy)*p_j_tilde(eq[2],eq[3])*eq[5])
theta_x0(eq) = (s_u*p_u_tilde(eq[2],eq[3])*eq[1]+(1-h_x)*s_u*p_u_tilde(eq[2],eq[3])*eq[4]+h_x*(T_x(h_x_bar)*s_jy+(1-T_x(h_x_bar))*s_jx)*p_j_tilde(eq[2],eq[3])*eq[4]+(1-h_y)*s_u*p_u_tilde(eq[2],eq[3])*eq[5]+h_y*(T_y(h_y_bar)*s_jx+(1-T_y(h_y_bar))*s_jy)*p_j_tilde(eq[2],eq[3])*eq[5])/(eq[2]+eq[3]+s_u*p_u_tilde(eq[2],eq[3])*eq[1]+(1-h_x)*s_u*p_u_tilde(eq[2],eq[3])*eq[4]+h_x*(T_x(h_x_bar)*s_jy+(1-T_x(h_x_bar))*s_jx)*p_j_tilde(eq[2],eq[3])*eq[4]+(1-h_y)*s_u*p_u_tilde(eq[2],eq[3])*eq[5]+h_y*(T_y(h_y_bar)*s_jx+(1-T_y(h_y_bar))*s_jy)*p_j_tilde(eq[2],eq[3])*eq[5])
theta_y(eq) = (1-phi)*eq[2] 
psi_x(eq) = theta_x(eq)+1+phi
psi_y(eq) = theta_y(eq)+1+phi


#competition
a = 0.1
p_u = 0.6
p_u_tilde(x,y) = p_u/(1+a*(x+y))
p_j = 0.58
p_j_tilde(x,y) = p_j/(1+a*(x+y))

parameters = Dict("s_u"=>s_u,"s_jx"=>s_jx,"s_jy"=>s_jy,"s_x"=>s_x,"s_y"=>s_y,"phi"=>phi,"M"=>M,"a"=>a,"p_u"=>p_u,"p_j"=>p_j,"h_x" =>h_x,"h_x_bar" =>h_x_bar,"h_y" =>h_y, "h_y_bar" =>h_y_bar)

#time
t_0 = 0
t_max = 100

#Relatedness
R_jp = 1/2*(1+phi)
R_siblings = (phi^2+(1-phi)^2*(1/4+1/(4*M)))+1
```

```julia
# Wild type matrix
function A(x,y,parameters) 
  
    s_u = parameters["s_u"]
    s_jx = parameters["s_jx"]
    s_jy = parameters["s_jy"]
    s_x = parameters["s_x"]
    s_y = parameters["s_y"]
    p_u = parameters["p_u"]
    p_j = parameters["p_j"]
    a = parameters["a"]
    M = parameters["M"]
    phi = parameters["phi"]
    h_x = parameters["h_x"]
    h_y = parameters["h_y"]
    h_x_bar = parameters["h_x_bar"]
    h_y_bar = parameters["h_y_bar"]
    p_u_tilde(x,y) = p_u/(1+a*(x+y))
    p_j_tilde(x,y) = p_j/(1+a*(x+y))
    h_x = parameters["h_x"]
    h_y = parameters["h_y"]
    h_x_bar = parameters["h_x_bar"]
    h_y_bar = parameters["h_y_bar"]
    
    return [s_u*(1-p_u_tilde(x,y)) 0 0 (1-h_x)*s_u*(1-p_u_tilde(x,y))+h_x*(T_x(h_x_bar)*s_jy+(1-T_x(h_x_bar))*s_jx)*(1-p_j_tilde(x,y)) (1-h_y)*s_u*(1-p_u_tilde(x,y))+h_y*(T_y(h_y_bar)*s_jx+(1-T_y(h_y_bar))*s_jy)*(1-p_j_tilde(x,y))
    ; s_u*p_u_tilde(x,y) (1-T_x(h_x_bar))*s_x T_y(h_y_bar)*s_x (1-h_x)*s_u*p_u_tilde(x,y)+h_x*(T_x(h_x_bar)*s_jy+(1-T_x(h_x_bar))*s_jx)*p_j_tilde(x,y) (1-h_y)*s_u*p_u_tilde(x,y)+h_y*(T_y(h_y_bar)*s_jx+(1-T_y(h_y_bar)*s_jy))*p_j_tilde(x,y)
    ; 0 T_x(h_x_bar)*s_y (1-T_y(h_y_bar))*s_y 0 0
    ; s_u*p_u_tilde(x,y)*b_x(0) (1-T_x(h_x_bar))*s_x*b_x(s_jx) T_y(h_y_bar)*s_x*b_x(s_jx) (1-h_x)*s_u*p_u_tilde(x,y)*b_x(0)+h_x*(T_x(h_x_bar)*s_jy+(1-T_x(h_x_bar))*s_jx)*p_j_tilde(x,y)*b_x(0) (1-h_y)*s_u*p_u_tilde(x,y)*b_x(0)+h_y*(T_y(h_y_bar)*s_jx+(1-T_y(h_y_bar))*s_jy)*p_j_tilde(x,y)*b_x(0)
    ; 0 T_x(h_x_bar)*s_y*b_y(s_jy) (1-T_y(h_y_bar))*s_y*b_y(s_jy) 0 0
    ];
end

# Mutant matrix
function Mutant(eq,parameters) 
    x = eq[2]
    y = eq[3]
    s_u = parameters["s_u"]
    s_jx = parameters["s_jx"]
    s_jy = parameters["s_jy"]
    s_x = parameters["s_x"]
    s_y = parameters["s_y"]
    p_u = parameters["p_u"]
    p_j = parameters["p_j"]
    a = parameters["a"]
    p_u_tilde(x,y) = p_u/(1+a*(x+y))
    p_j_tilde(x,y) = p_j/(1+a*(x+y))
    M = parameters["M"]
    phi = parameters["phi"]
    h_x = parameters["h_x"]
    h_y = parameters["h_y"]
    h_x_bar = parameters["h_x_bar"]
    h_y_bar = parameters["h_y_bar"]
    return [s_u*(1-p_u_tilde(x,y)) 0 0 (1-h_x)*s_u*(1-p_u_tilde(x,y))+h_x*(T_x(h_x_bar)*s_jy+(1-T_x(h_x_bar))*s_jx)*(1-p_j_tilde(x,y)) (1-h_y)*s_u*(1-p_u_tilde(x,y))+h_y*(T_y(h_y_bar)*s_jx+(1-T_y(h_y_bar))*s_jy)*(1-p_j_tilde(x,y))
    ; s_u*p_u_tilde(x,y) (1-T_x(h_x_bar))*s_x T_y(h_y_bar)*s_x (1-h_x)*s_u*p_u_tilde(x,y)+h_x*(T_x(h_x_bar)*s_jy+(1-T_x(h_x_bar))*s_jx)*p_j_tilde(x,y) (1-h_y)*s_u*p_u_tilde(x,y)+h_y*(T_y(h_y_bar)*s_jx+(1-T_y(h_y_bar))*s_jy)*p_j_tilde(x,y)
    ; 0 T_x(h_x_bar)*s_y (1-T_y(h_y_bar))*s_y 0 0
        
    ; s_u*p_u_tilde(x,y)*(b_x(0)/2*(1+phi+(1-phi)*theta_x0(eq))+b_x(s_jx)/2*theta_x(eq)) (1-T_x(h_x_bar))*s_x*b_x(s_jx)/2*psi_x(eq)+T_x(h_x_bar)*s_y*b_x(s_jx)/2*theta_x(eq)+((1-T_x(h_x_bar))*s_x+T_x(h_x_bar)*s_y)*b_0/2*theta_x0(eq) (1-T_y(h_y_bar))*s_y*b_y(s_jy)/2*psi_y(eq)+T_y(h_y_bar)*s_x*b_y(s_jy)/2*theta_y(eq)+((1-T_y(h_y_bar))*s_y+T_y(h_y_bar)*s_x)*b_0/2*theta_x0(eq) (1-h_x)*s_u*p_u_tilde(x,y)*b_x(0)/2*(1+phi+theta_x0(eq))+(1-h_x)*s_u*p_u_tilde(x,y)*b_x(s_jx)/2*theta_x(eq)+h_x*(T_x(h_x_bar)*s_jy+(1-T_x(h_x_bar))*s_jx)*p_j_tilde(x,y)*b_x(0)/2*(1+phi+theta_x0(eq))+h_x*(T_x(h_x_bar)*s_jy+(1-T_x(h_x_bar))*s_jx)*p_j_tilde(x,y)*b_x(s_jx)/2*theta_x(eq) (1-h_y)*s_u*p_u_tilde(x,y)*b_x(0)/2*(1+phi+theta_x0(eq))+(1-h_y)*s_u*p_u_tilde(x,y)*b_x(s_jx)/2*theta_x(eq)+h_y*(T_y(h_y_bar)*s_jx+(1-T_y(h_y_bar))*s_jy)*p_j_tilde(x,y)*b_x(0)/2*(1+phi+theta_x0(eq))+h_y*(T_y(h_y_bar)*s_jx+(1-T_y(h_y_bar))*s_jy)*p_j_tilde(x,y)*b_x(s_jx)/2*theta_x(eq)
    ; s_u*p_u_tilde(x,y)*b_y(s_jy)/2*theta_y(eq) (1-T_x(h_x_bar))*s_x*b_y(s_jy)/2*theta_y(eq)+T_x(h_x_bar)*s_y*b_y(s_jy)/2*psi_y(eq) T_y(h_y_bar)*s_x*b_y(s_jy)/2*theta_y(eq)+(1-T_y(h_y_bar))*s_y*b_y(s_jy)/2*psi_y(eq) (1-h_x)*s_u*p_u_tilde(x,y)*b_y(s_jy)/2*theta_y(eq)+h_x*(T_x(h_x_bar)*s_jy+(1-T_x(h_x_bar))*s_jx)*p_j_tilde(x,y)*b_y(s_jy)/2*theta_y(eq) (1-h_y)*s_u*p_u_tilde(x,y)*b_y(s_jy)/2*theta_y(eq)+h_y*(T_y(h_y_bar)*s_jx+(1-T_y(h_y_bar))*s_jy)*p_j_tilde(x,y)*b_y(s_jy)/2*theta_y(eq)
    ];
end;

function equilibrium_mu(parameters)
    mu = rand!(zeros(5,1))
    while sum(abs.(mu - A(mu[2], mu[3],parameters)*mu)) >0.00001
        mu = A(mu[2], mu[3],parameters)*mu
    end
    return mu
end;

function equilibrium_nu(parameters,mu)
    nu = rand!(zeros(1,5))
    while sum(abs.(nu - nu*A(mu[2], mu[3],parameters))) >0.00001
        nu = nu*A(mu[2], mu[3],parameters)/sum(nu)
    end
    return nu
end;


function R_0(parameters)

    s_u = parameters["s_u"]
    s_jx = parameters["s_jx"]
    s_jy = parameters["s_jy"]
    s_x = parameters["s_x"]
    s_y = parameters["s_y"]
    p_u = parameters["p_u"]
    p_j = parameters["p_j"]
    a = parameters["a"]
    p_u_tilde(x,y) = p_u/(1+a*(x+y))
    p_j_tilde(x,y) = p_j/(1+a*(x+y))
    M = parameters["M"]
    phi = parameters["phi"]
    h_x = parameters["h_x"]
    h_y = parameters["h_y"]
    h_x_bar = parameters["h_x_bar"]
    h_y_bar = parameters["h_y_bar"]
    nu_u = 1
    nu_jx = (1 - h_x) + h_x*((1 − T_x(h_x_bar))*s_jx + T_x(h_x_bar)*s_jy )*(1 − p_j+p_j*(1-s_u*(1-p_u))/s_u/p_u)
    nu_jy = (1 - h_y) + h_y*((1 − T_y(h_y_bar))*s_jy + T_y(h_y_bar)*s_jx )*(1 − p_j+p_j*(1-s_u*(1-p_u))/s_u/p_u)
    nu_y = 1/(1-s_y*(1-T_y(h_y_bar)))*(((1-T_y(h_y_bar))*s_y*b_y(s_jy)*nu_jy)+T_y(h_y_bar)*s_x*(1-s_u*(1-p_u))/s_u/p_u)
    nu_x = (1-T_x(h_x_bar))*s_x*(1-s_u*(1-p_u))/s_u/p_u+T_x(h_x_bar)*s_y*(nu_y+b_y(s_jy)*nu_jy)
    R_0 = s_u*(1-p_u)*nu_u+s_u*p_u*(nu_x+b_x(0)*nu_jx)
    return R_0
end
```

```julia
function Fitness_comput(parameters)
    d_hx = 0.01
    d_hx_bar = 0.01
    d_hy = 0.01
    d_hy_bar = 0.01
    d_sjx = 0.01
    d_sjy = 0.01
    dW_x = 0
    dW_y = 0
    # Find eigenvectors: mu, nu
    if R_0(parameters)>1.05
        mu = equilibrium_mu(parameters)
        nu = equilibrium_nu(parameters, mu)

        # h_x, h_x_bar
        parameters["h_x"] += d_hx
        W_indiv_hx_up = nu*Mutant(mu[2], mu[3],parameters)*[0,0,0,b_x(s_jx)*mu[2],0]
        parameters["h_x"] -= 2*d_hx
        W_indiv_hx_down = nu*Mutant(mu[2], mu[3],parameters)*[0,0,0,b_x(s_jx)*mu[2],0]

        parameters["h_x"] += d_hx
        parameters["h_x_bar"] +=d_hx_bar
        # changing right eigenvector
        W_parent_hx_up = nu*Mutant(mu[2], mu[3],parameters)*[0,mu[2],0,0,0]
        #To be discussed:
        #parameters["h_x"] += d_hx
        W_siblings_hx_up = nu*Mutant(mu[2], mu[3],parameters)*[0,0,0,b_x(s_jx)*mu[2],0]
        #parameters["h_x"] -= d_hx
        
        parameters["h_x_bar"] -= 2*d_hx_bar
        W_parent_hx_down = nu*Mutant(mu[2], mu[3],parameters)*[0,mu[2],0,0,0]
        #To be discussed:
        #parameters["h_x"] -= *d_hx
        W_siblings_hx_down = nu*Mutant(mu[2], mu[3],parameters)*[0,0,0,b_x(s_jx)*mu[2],0]
        #parameters["h_x"] += d_hx
        
        parameters["h_x_bar"] +=d_hx_bar


        W_indiv_hx = W_indiv_hx_up-W_indiv_hx_down
        W_parent_hx = R_jp*(W_parent_hx_up-W_parent_hx_down)
        W_siblings_hx = R_siblings*(W_siblings_hx_up-W_siblings_hx_down)
        dW_x = W_parent_hx[1]+W_indiv_hx[1]+W_siblings_hx[1]
        
        #s_jx change
        parameters["s_jx"] += d_sjx
        W_indiv_sjx_up = nu*Mutant(mu[2], mu[3],parameters)*[0,0,0,b_x(s_jx)*mu[2],0]
        W_parent_sjx_up = nu*Mutant(mu[2], mu[3],parameters)*[0,mu[2],0,0,0]
        parameters["s_jx"] -= 2*d_sjx
        W_indiv_sjx_down = nu*Mutant(mu[2], mu[3],parameters)*[0,0,0,b_x(s_jx)*mu[2],0]
        W_parent_sjx_down = nu*Mutant(mu[2], mu[3],parameters)*[0,mu[2],0,0,0]
        W_indiv_sjx = W_indiv_sjx_up-W_indiv_sjx_down
        W_parent_sjx = R_jp*(W_parent_sjx_up-W_parent_sjx_down)
        dW_sjx = W_indiv_sjx[1]+W_parent_sjx[1]
        
        
        # h_y, h_y_bar

        parameters["h_y"] += d_hy
        W_indiv_hy_up = nu*Mutant(mu[2], mu[3],parameters)*[0,0,0,0,b_y(s_jy)*mu[3]]
        parameters["h_y"] -= 2*d_hy
        W_indiv_hy_down = nu*Mutant(mu[2], mu[3],parameters)*[0,0,0,0,b_y(s_jy)*mu[3]]

        parameters["h_y"] += d_hy
        parameters["h_y_bar"] += d_hy_bar
        W_parent_hy_up = nu*Mutant(mu[2], mu[3],parameters)*[0,0,mu[3],0,0]
        W_siblings_hy_up = nu*Mutant(mu[2], mu[3],parameters)*[0,0,0,0,b_y(s_jy)*mu[3]]

        parameters["h_y_bar"] -= 2*d_hy_bar
        W_parent_hy_down = nu*Mutant(mu[2], mu[3],parameters)*[0,0,mu[3],0,0]
        W_siblings_hy_down = nu*Mutant(mu[2], mu[3],parameters)*[0,0,0,0,b_y(s_jy)*mu[3]]
        parameters["h_y_bar"] += d_hy_bar


        W_indiv_hy = W_indiv_hy_up-W_indiv_hy_down
        W_parent_hy = R_jp*(W_parent_hy_up-W_parent_hy_down)
        W_siblings_hy = R_siblings*(W_siblings_hy_up-W_siblings_hy_down)
        
        dW_y = W_parent_hy[1]+W_indiv_hy[1]+W_siblings_hy[1]
        
        #s_jy change
        parameters["s_jy"] += d_sjy
        W_indiv_sjy_up = nu*Mutant(mu[2], mu[3],parameters)*[0,0,0,0,b_y(s_jy)*mu[3]]
        W_parent_sjy_up = nu*Mutant(mu[2], mu[3],parameters)*[0,0,mu[3],0,0]
        parameters["s_jy"] -= 2*d_sjy
        W_indiv_sjy_down = nu*Mutant(mu[2], mu[3],parameters)*[0,0,0,0,b_y(s_jy)*mu[3]]
        W_parent_sjy_down = nu*Mutant(mu[2], mu[3],parameters)*[0,0,mu[3],0,0]
        W_indiv_sjy = W_indiv_sjy_up-W_indiv_sjy_down
        W_parent_sjy = R_jp*(W_parent_sjy_up-W_parent_sjy_down)
        dW_sjy = W_indiv_sjy[1]+W_parent_sjy[1]
    end  
    return dW_x, dW_y, dW_sjx, dW_sjy
   
    
end;

```

```julia
function ESS(parameters)
    dW_x = 1
    dW_y = 1
    dW_sjx = 1
    dW_sjy = 1
    i=0
    h_x_mem = 1
    h_y_mem = 1
    sjx_mem = 1
    sjy_mem = 1
    reason = "ess found"
    while abs(dW_x) > 1e-5 || abs(dW_y) > 1e-5 ||abs(dW_sjx) > 1e-5 || abs(dW_sjy) > 1e-5 
      
        i += 1
        #If taking too much time
        if i > 20000
            reason = "i>100000"
            break
        end
        
        
        #Same values previously computed
        if parameters["h_x"]==h_x_mem && parameters["h_y"]==h_y_mem && parameters["s_jx"]==sjx_mem && parameters["s_jy"]==sjy_mem
            reason = "stagnation"
            break
        end
        
        if R_0(parameters)<1.02
            break
        end
        
        dW_x, dW_y, dW_sjx, dW_sjy = Fitness_comput(parameters)
        
        #Loading previous values
        h_x_mem = parameters["h_x"]
        h_y_mem = parameters["h_y"]
        sjx_mem = parameters["s_jx"]
        sjy_mem = parameters["s_jy"]
        #New values
        parameters["h_x"] += dW_x
        parameters["h_x_bar"] += dW_x
        parameters["s_jx"] += dW_sjx
        parameters["h_y"] += dW_y
        parameters["h_y_bar"] += dW_y
        parameters["s_jy"] += dW_sjy
        
        #Boundaries conditions
        parameters["h_x"] = clamp(parameters["h_x"],0,1)
        parameters["h_y"] = clamp(parameters["h_y"],0,1)
        parameters["h_x_bar"] = parameters["h_x"]
        parameters["h_y_bar"] = parameters["h_y"]
        parameters["s_jx"] = clamp(parameters["s_jx"],0,1)
        parameters["s_jy"] = clamp(parameters["s_jy"],0,1)

    end
    parameters["h_x"] = clamp(parameters["h_x"],0,1)
    parameters["h_y"] = clamp(parameters["h_y"],0,1)
    parameters["h_x_bar"] = parameters["h_x"]
    parameters["h_y_bar"] = parameters["h_y"]
    parameters["s_jx"] = clamp(parameters["s_jx"],0,1)
    parameters["s_jy"] = clamp(parameters["s_jy"],0,1)
    return parameters

end
```
